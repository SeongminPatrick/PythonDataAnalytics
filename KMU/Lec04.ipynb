```python
import pandas
```

```python
automobile = pandas.read_csv('automobile.csv')
```

```python
import seaborn
```

```python
%matplotlib inline
```

```python
seaborn.regplot(automobile.length, automobile.symboling)
```

```python
seaborn.regplot('length', 'symboling', data=automobile)
```

```python
seaborn.regplot('length', 'symboling', data=automobile, fit_reg=False)
```

```python
seaborn.regplot('length', 'symboling', data=automobile, fit_reg=False, y_jitter=0.1)
```

```python
seaborn.regplot('length', 'symboling', data=automobile, y_jitter=0.1)
```

```python
seaborn.regplot('length', 'symboling', data=automobile, y_jitter=0.1, ci=None, line_kws={'color': 'red'}, scatter_kws={'color': 'orange'})
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols('symboling ~ length', data=automobile)
```

```python
result = model.fit()
```

```python
result.summary()
```
