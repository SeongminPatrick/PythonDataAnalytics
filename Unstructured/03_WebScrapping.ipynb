## HTML 처리

```python
import lxml.html
```

```python
root = lxml.html.fromstring(response.text)
root
```

```python
for node in root.xpath('.//a[@class="tit"]'):
    print(node.text_content(), node.attrib['href'])
```

## 목록에서 여러 개의 링크 가져오기

```python
res = requests.get('https://techcrunch.com/startups/')
res
```

```python
root = lxml.html.fromstring(res.text)
```

```python
titles = root.cssselect('h2 a')   # 기사 제목
titles
```

```python
for node in root.xpath('.//a[@class="post-title"]'):
    print(node.text_content(), node.attrib['href'])
```

```python
for title in titles:
    print(title.text, title.attrib['href'])
```

```python
articles = []
for title in titles:
    url =  title.attrib['href'];
    res = requests.get(url);
    articles.append(res.text)
```
